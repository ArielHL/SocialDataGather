```python
import json
from client import RestClient
import pandas as pd
# You can download this file from here https://cdn.dataforseo.com/v3/examples/python/python_Client.zip
client = RestClient("ariel.hernan.limes@pwc.com", "be0a9375b921d1dd")

function ='asin'


# get the task results by id
# GET /v3/merchant/amazon/products/task_get/advanced/$id
id = "09281817-6489-0192-0000-b9434eb1f514"
response = client.get(f"/v3/merchant/amazon/{function}/task_get/advanced/" + id)
# you can find the full list of the response codes here https://docs.dataforseo.com/v3/appendix/errors
if response["status_code"] == 20000:
    with open(f"./Task-Get/task_get_by_id_{id}.json", "w") as outfile:
        json.dump(response, outfile, indent=4)
    # do something with result
else:
    print("error. Code: %d Message: %s" % (response["status_code"], response["status_message"])) 
```

```python
# reading json file
get_result=json.load(open('task_get_by_id.json'))
# get item list
item_list=get_result['tasks'][0]['result'][0]['items']

# Create DataFrame
df=pd.DataFrame(item_list)

# expand dataframe with rating data
df= pd.concat([df,df['rating'].apply(pd.Series)],axis=1)
# expand dataframe with info data
df= pd.concat([df,df['delivery_info'].apply(lambda x: pd.Series(x, dtype=object))], axis=1)
df.drop(['rating','delivery_info','xpath','items'],axis=1,inplace=True)
```
